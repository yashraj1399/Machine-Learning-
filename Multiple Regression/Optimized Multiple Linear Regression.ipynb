```python
#Using All The Previous Code for The X and Y 
import pandas as pd
import numpy as  np
import matplotlib.pyplot as plt
dataset=pd.read_csv("50_Startups.csv")
X=dataset.iloc[:,:-1].values
Y=dataset.iloc[:,4].values
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
labelencoder_X=LabelEncoder()
X[:,3]=labelencoder_X.fit_transform(X[:,3])
onehotencoder=OneHotEncoder(categorical_features=[3])
X=onehotencoder.fit_transform(X).toarray()
X=X[:,1:]
```

```python
#Now we will import the new Library named statsmodel for optimization of the Multiple Linear Regression Solution 
```

```python
import statsmodels.formula.api as sm
```

```python
#Adding the ones in the first column of the matrix which we have created as if don't add this then our library wouldn't consider the b0 cons
#tant as equation of Multiple Linear Regreesion is y=b0*x0+b1*x1+b2*x2 .....
X=np.append(arr=np.ones((50,1)).astype(int),values=X,axis=1)
X_opt=X[:,[0,1,2,3,4,5]]
```

```python
regressor_OLS=sm.OLS(endog=Y,exog=X_opt).fit()
regressor_OLS.summary()

#Here For each step we have to pop out the largest P value from the Matrix and Implment this Operation Again
```

```python
X_opt=X_opt=X[:,[0,1,3,4,5]]
regressor_OLS=sm.OLS(endog=Y,exog=X_opt).fit()
regressor_OLS.summary()
```

```python
X_opt=X_opt=X[:,[0,3,4,5]]
regressor_OLS=sm.OLS(endog=Y,exog=X_opt).fit()
regressor_OLS.summary()
```

```python
X_opt=X_opt=X[:,[0,3,5]]
regressor_OLS=sm.OLS(endog=Y,exog=X_opt).fit()
regressor_OLS.summary()
```

```python
X_opt=X_opt=X[:,[0,3]]
regressor_OLS=sm.OLS(endog=Y,exog=X_opt).fit()
regressor_OLS.summary()
```
